## Step 2: Configure Environment Variables


```python
import os

# Required: Set your dorc-engine Cloud Run service URL
os.environ["DORC_BASE_URL"] = "https://your-engine-url.run.app"

# Required: Set your tenant identifier
os.environ["DORC_TENANT_SLUG"] = "my-tenant"

# Optional: Set API key if authentication is required
# os.environ["DORC_API_KEY"] = "your-api-key-here"

print("Environment variables configured:")
print(f"  DORC_BASE_URL: {os.environ.get('DORC_BASE_URL')}")
print(f"  DORC_TENANT_SLUG: {os.environ.get('DORC_TENANT_SLUG')}")
print(f"  DORC_API_KEY: {'***' if os.environ.get('DORC_API_KEY') else 'Not set'}")
```

## Step 3: Test Connection


```python
from dorc_client import DorcClient

# Create client
client = DorcClient()

# Check health
print("Checking dorc-engine health...")
is_healthy = client.health()

if is_healthy:
    print("✅ dorc-engine is healthy and ready!")
else:
    print("❌ dorc-engine is not responding. Check your DORC_BASE_URL.")
```
